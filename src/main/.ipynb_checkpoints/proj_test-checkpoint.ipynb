```python
data["text"] = data['headline'].map(str) + " " + data['short_description'].map(str)
X = data["text"].apply(str)
X.values
```

```python
min_recs = min(data.groupby('category')['text'].nunique())
print("Smallest sample in category: ", min_recs)
data.groupby('category')['text'].nunique()
```

```python
data.loc[data['category'] == 'ARTS', 'category'] = 'ARTS & CULTURE'
data.loc[data['category'] == 'CULTURE & ARTS', 'category'] = 'ARTS & CULTURE'
data.loc[data['category'] == 'STYLE', 'category'] = 'STYLE & BEAUTY'
```

```python
data.groupby('category')['text'].nunique()
```

```python
atts = ['SCIENCE', 'RELIGION', 'TECH', 'ENVIRONMENT', 'HOME & LIVING', 'SPORTS', 'BUSINESS', 'TRAVEL', 'STYLE & BEAUTY', 'ENTERTAINMENT', 'POLITICS', 'FOOD & DRINK', 'CRIME', 'ARTS & CULTURE', 'COMEDY', 'WORLD NEWS', 'EDUCATION' ]
for x in atts:
    if x not in data['category'].values:
        print(x)
    else:
        print(x + ": GOOD to go!")
```

```python
data_sub = data.loc[data['category'].isin(atts)]
print(data_sub.groupby('category')['text'].nunique())
```

```python
min_recs_sub = min(data_sub.groupby('category')['text'].nunique())
print("Smallest sample in category: ", min_recs_sub)
```

```python
balanced_data = data_sub.groupby('category', as_index=False, group_keys=False).apply(lambda s: s.sample(min_recs_sub +10000,replace=True))
print(len(balanced_data))
X = balanced_data['text']
y = balanced_data['category']
balanced_data.groupby('category')['text'].count()   # no longer unique
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size =0.2, random_state=0)
```

```python
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.fit_transform(y_test)
#print(le.classes_)
#len(le.classes_)
#y = le.transform(y)
#y
#X.values
print(y_train_enc)
print(le.inverse_transform(y_train_enc))
```

```python
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(X)

# transform the training and validation data using count vectorizer object
X_train_count =  count_vect.transform(X_train)
X_test_count =  count_vect.transform(X_test)
```

```python
# word level tf-idf
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=100000)
tfidf_vect.fit(X)
X_train_tfidf =  tfidf_vect.transform(X_train)
X_test_tfidf =  tfidf_vect.transform(X_test)
#input_text = tfidf_vect.transform(["Kawhi Leonard, the NBA's Kingslayer, on the Verge of Taking Down Another Dynasty"])
```

```python
clfr_NB_count = MultinomialNB(alpha = 0.05)
clfr_NB_count.fit(X_train_count, y_train_enc)
preds = clfr_NB_count.predict(X_test_count)
print(preds)
print(y_test_enc)
print()
score = accuracy_score(y_test_enc, preds)
labels = balanced_data['category'].unique().tolist()
cm = confusion_matrix(le.inverse_transform(y_test_enc), le.inverse_transform(preds), labels=balanced_data['category'].unique().tolist())
print(score)
print()
print()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
clfr_NB_tf = MultinomialNB(alpha = 0.05)
clfr_NB_tf.fit(X_train_tfidf, y_train_enc)

preds = clfr_NB_tf.predict(X_test_tfidf)
print(preds)
print(y_test_enc)
print()
score = accuracy_score(y_test_enc, preds)
labels = balanced_data['category'].unique().tolist()
cm = confusion_matrix(le.inverse_transform(y_test_enc), le.inverse_transform(preds), labels=balanced_data['category'].unique().tolist())
print(score)
print()
print()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
# Linear Classifier on Count Vectors
clfr_LR_count = linear_model.LogisticRegression(multi_class='auto')
clfr_LR_count.fit(X_train_count, y_train_enc)
preds = clfr_LR_count.predict(X_test_count)
print(preds)
print(y_test_enc)
print()
score = accuracy_score(y_test_enc, preds)
print(score)
cm = confusion_matrix(y_test_enc, preds, le.labels)
print(score)
print()
print(cm)
```

```python
# Linear Classifier on tfIDF Vectors
clfr_LR_tf = linear_model.LogisticRegression(multi_class='auto', solver='lbfgs')
clfr_LR_tf.fit(X_train_tfidf, y_train_enc)
preds = clfr_LR_tf.predict(X_test_tfidf)
print(preds)
print(y_test_enc)
print()
score = accuracy_score(y_test_enc, preds)
labels = balanced_data['category'].unique().tolist()
cm = confusion_matrix(le.inverse_transform(y_test_enc), le.inverse_transform(preds), labels=balanced_data['category'].unique().tolist())
print(score)
print()
print()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
